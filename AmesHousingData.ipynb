##**Categorical data**

From below graph in step 28, we could say that most preferred house styles to be built during the recent years were of 1 and 2 story styles however in the 90's, mostly houses of 2.5 Fin to 2.5 Unf were built.

```python
# Step 28: Visualizing categorical data using boxplot
sns.boxplot(x=ames3["Year Built"], y=ames3["House Style"])
```

From the below analysis in step 29 and 30, we could see that there is a major disparity in the sales price among ames neighbourhood houses. StoneB,NridgHt and NoRidge are the 3 most costly area with median sales price of around $300,000 which is 3 times as high as the median of the 3 least expensive areas, which are MeadowV, DOTRR and BrDale.

```python
# Step 29: Visualize categorical feature of property connected with saleprice
def visualizeCatFeature(feature):
    """
    Visualize categorical feature of property connected with saleprices 
    """
    # Arranging median saleprice in decending order 
    sale = ames3.groupby(
        [feature]).median().SalePrice.sort_values(ascending=False).index

    # Creating Box Plot
    sns.boxplot(x=feature,
                y='SalePrice',
                data=ames3,
                order=sale,
                palette='BrBG')
```

```python
# Step 30: Visualizing SalesPrice based on Neighborhood using CatFeature
plt.figure(figsize=(10, 5))
visualizeCatFeature('Neighborhood')
heading = plt.title('Sales Price based on Neighborhood')
# Rotating the Text of Xticks Label 
kit = plt.xticks(rotation=45)
```

From the below analysis in step 31, we can see that most costly houses are made of Wood Shingles rooftop which costs between 300,000 to 450,000. However, there are many houses which are costly whose rooftop is made of composite Shingle.

```python
# Step 31: Visualize Sales Price of a House based on Roof Material
# Set the figure size
plt.figure(figsize=(7, 5))
visualizeCatFeature('Roof Matl')
# Set the title of image
heading = plt.title('Sales Price of a House based on Roof Material')
```

As shown in step 32, we have visualized the correlation between multiple columns of ames dataset like YearBuilt and GarageYrBlt are correlated with each other which indicates that carports and houses were built together.
Likewise we have also seen that correlation between KitchenQual and ExteriorQual which shows that the new houses built, most likely has a good kitchen and exterior. 

```python
# Step 32: Visualizing the correlation among columns 
# Computing the correlation matrix
corr = ames3[['SalePrice'] + numericalColumn].corr()
fig = plt.figure(figsize=(16,15))
ax = fig.add_subplot(111)
# To generate custom diverging color map
cmap = sns.diverging_palette(120, 10, as_cmap=True)
# Drawing the heatmap 
sns.heatmap(corr, 
                 yticklabels=corr.index.values,
                 xticklabels=corr.columns.values,
           cmap=cmap)
ax.xaxis.tick_top()
# Rotating the Text of Xticks Label 
plt.setp(ax.get_xticklabels(), rotation=90)
# Display all the figures
plt.show()
```
